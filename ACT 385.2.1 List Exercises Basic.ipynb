#**Challenge: The "VIP Guest List" Manager**
###**Scenario**
You are managing the guest list for a high-profile tech event. You start with a basic list, but as the event approaches, several changes happen. Your task is to manipulate the list using Python to reach the final correct state.

**The Initial List**

````python
# Initial guest list
guests = ["Alice", "Bob", "Charlie", "David", "Eve"]
print("Initial list:", guests)
````

## **Your Tasks**

- **Late Arrival:**  
  A new guest, **"Frank"**, arrives. Add him to the **end of the list**.

- **VIP Priority:**  
  A VIP guest, **"Grace"**, needs to be added to the **very front (index 0)** of the list.

- **The Substitution:**  
  **"Charlie"** cannot make it. Replace his name with **"Chuck"**.

- **Security Issue:**  
  The person at **index 3** is on the "no-fly" list. Remove them from the list **by their name**.

- **Final Count:**  
  Print the **final list** and the **total number of guests currently attending**.

````python
# Late Arrival: Add Frank to the end
guests = ["Alice", "Bob", "Charlie", "David", "Eve"]
guests.append("Frank")
print("Initial list:", guests)
````

````python
# VIP Priority: Add Grace to the front
guests = ["Alice", "Bob", "Charlie", "David", "Eve"]
guests.insert (0, "Grace")
print("initial list:", guests)
````

````python
# The Substitution: Replace Charlie with Chuck
guests = ["Alice", "Bob", "Charlie", "David", "Eve"]
charlie_index = guests.index("Charlie")
guests[charlie_index] = "Chuck"
print("initial list:", guests)
````

````python
# Security Issue: Remove the person currently at index 3 by their name
person_to_remove = guests[3]
guests.remove(person_to_remove)
````

````python
# Final Count
print("Final guest list:", guests)
print("Total guests attending:", len(guests))
````
